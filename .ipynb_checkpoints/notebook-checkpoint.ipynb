```python
import pandas as pd
tourist_data = pd.read_excel('turista.xlsx')
```

```python
tourist_data.columns
```

```python
tourist_data.shape
```

```python
features = tourist_data.iloc[:, -19:-1]
features.head()
```

```python
features = features.replace({
    'Molto': 1,
    'Abbastanza': 0.66,
    'Poco': 0.33,
    'Per niente': 0
})
features.head()
```

```python
## Build a confidence interval for each column mean
from scipy.stats import t

confidence = 0.95
n = features.shape[0]
t_statistic = t.ppf((1 + confidence) / 2, n - 1)

means = features.mean()
standard_errors = features.sem()
confidence_intervals = pd.DataFrame({
    'mean': means,
    'lower_bound': means - t_statistic * standard_errors,
    'upper_bound': means + t_statistic * standard_errors
}).sort_values('mean', ascending=False)
```

```python
confidence_intervals.T.boxplot(rot=90, figsize=(20, 8));
```

```python
scores = features.mean().sort_values(ascending=False).iloc[:5]
```

```python
scores / scores.sum()
```

```python
scores_by_variance = confidence_intervals['mean'] / (confidence_intervals['upper_bound'] - confidence_intervals['lower_bound'])
scores_by_variance = scores_by_variance.sort_values(ascending=False)
scores_by_variance[:5] / scores_by_variance[:5].sum()
```

```python


# Hard constraint filter

tour = {
    # Hard constraints
    'need_tickets': True,
    'group': True,

    # Soft constraints
    'attractions': [('city', 'attraction')],

    # Information
    'max_people': 20,
    'min_people': 5, 
    'cost': 5,
    'duration_time': 3,
    'duration_distance': 5,
    'accessiblity': True
}

guide = {
    # Hard constraints
    'tour_types': ['Tour'],
    'languages' : ['Italian'],
    'availability': False, # Specified by the guide

    # Soft constraints
    'ratings': 4.5,
    'experience': 4, # Number of years worked
    'sex': 'Male',
    'cost_per_hour': 15,
    'attractions': [('Rome','Colosseum')],

    # Information
    'name': 'Lorenzo',
    'picture': 0,
    'age': 27,
    'education': '',
    'tour': [tour]
}

guide2 = {
    # Hard constraints
    'tour_types': ['Sightseeing'],
    'languages' : ['Italian', 'English'],
    'availability': True, # Specified by the guide

    # Soft constraints
    'ratings': 4.3,
    'experience': 4, # Number of years worked
    'sex': 'Male',
    'cost_per_hour': 20,
    'attractions': [('Paris', 'Eiffel Tower')],

    # Information
    'name': 'Lorenzo',
    'picture': 0,
    'age': 27,
    'education': '',
    'tour': [tour]
}


hard_constraints = {
    'languages': ['Italian'],
    'tour_type': ['Sightseeing']
}

guide_data = [guide, guide2]

def filter_guide(hard_constraints, guide):
    
    if guide['availability'] == False:
        return False
    
    if len(set.intersection(set(hard_constraints['languages']), set(guide['languages']))) == 0 :
        return False
    
    if len(set.intersection(set(hard_constraints['tour_type']), set(guide['tour_types']))) == 0 :
        return False
    
    return True

def hard_constraint_filter(hard_constraints, guide_data):
    
    res = list(filter(lambda guide: filter_guide(hard_constraints, guide), guide_data))
    return res


print(hard_constraint_filter(hard_constraints, guide_data))
```
